```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from time import time
from keras.models import Sequential, load_model
from keras.layers import GRU, LSTM, Dense
from keras.optimizers import Adam

from sklearn.preprocessing import MinMaxScalar
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
path = ""
data = pd.read_csv(path)
data.sort_values(by="Date", inplace=True)
data = data.reset_index(drop=True)
data = data[data['Word'].str.len() == 5]
data = data[data["Number of the reported results"] > 10000]

data_x = np.zeros(shape=(len(data), 5, 26))

train_ratio = 0.8
train_x, test_x, train_y, test_y = train_test_plit(data_x, data_y, train_size=train_ratio)

model = Sequential()
model.add(GRU(512, input_shape=(5, 26)))
model.add(Dense(7))

model.compile(loss='mse', optimizer=Adam(learning_rate=1e-4))

start_time = time()
model.fit(train_x.reshape(-1, 5, 26), train_y, epochs=1000, batch_size=1, verbose=2)
print("Training time : ", time() - start_time)

model.save('GRU.h5')
predictions = model.predict(test_x.reshape(-1, 5, 26))
print(model.count_params())
```
